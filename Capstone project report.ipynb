# Methodology

After cleaning up and preparing our data, they are now ready to be fed into machine learning models. We will be mainly using the Decision Tree method. A decision tree model gives us a layout of all possible outcomes so we can fully analyze the consequences of a decision. Under this context, the decision tree observes all possible outcomes of different junction types and different weather conditions. # Results

As seen from the predicted results under the prediction section earlier, this Decision Tree model that I chose to use is not fully accurate and does have some flaws present. But that is with all prediction models, no model can be fully forsee the results with 100% accruacy. However, it is commendable and this model received a Accuracy classification score of 0.7380503402476019. 

Accuracy classification score computes subset accuracy: the set of labels predicted for a sample must exactly match the corresponding set of labels in y_true. In multilabel classification, the function returns the subset accuracy. If the entire set of predicted labels for a sample strictly match with the true set of labels, then the subset accuracy is 1.0; otherwise it is 0.0.

Therefore, the score of 0.7380503402476019 for our model is already considered to be very good in predicting outcome of accident at different junctions and weather conditions, which is what our analysis aims to achieve.
